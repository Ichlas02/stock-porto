# Add code field to each stock data & Trimming IHSG to LQ45

```python
import pandas as pd
import os
import shutil
import tvDatafeed as tvd
from dotenv import load_dotenv
from pathlib import Path
```

```python
# List of LQ45 stock codes
lq45_codes = ["AALI", "ADHI", "ADRO", "ADMR", "AKRA", "ANTM", "ASII", "ASRI", "BBCA",
               "BBNI", "BBRI", "BBTN", "BMRI", "BMTR", "BSDE", "CPIN", "GGRM", "HMSP",
               "ICBP", "INCO", "INDF", "INTP", "JSMR", "KLBF", "LPKR", "LPPF", "LSIP",
               "MNCN", "MPPA", "MYRX", "PGAS", "PTBA", "PTPP", "PWON", "SCMA", "SILO",
               "SMGR", "SMRA", "SRIL", "SSMS", "TBIG", "TLKM", "UNTR", "UNVR", "WSKT"]
```

```python
# Directory where your CSV files are stored
lq45_path = "../datasets/LQ45"

# Loop through each stock code
for code in lq45_codes:
    file_path = os.path.join(lq45_path, f"{code}.csv")
    
    if os.path.exists(lq45_path):
        # Check if the file exists
        if os.path.exists(file_path):
            # Read CSV file
            df = pd.read_csv(file_path)
            
            # Add "Code" column
            df["Code"] = code
            
            # Copy the file to the LQ45 folder
            shutil.copy(file_path, os.path.join(lq45_path, f"{code}.csv"))

            # Rewrite CSV
            df.to_csv(file_path, index=False)
        else:
            print(f"{code}.csv not found.")
    else:
        # Create LQ45 directory if it doesn't exist
        os.makedirs(lq45_path, exist_ok=True)
```
